##Correlation Analysis

- Correlation analysis is used to examine the relationship between numerical variables in the dataset.
- It helps identify whether changes in one variable are associated with changes in another.

##Movies and TV Shows Distribution

-This visualization compares the number of Movies and TV Shows available on Netflix, providing an understanding of the platform's content composition.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("netflix_titles.csv")

count = df['type'].value_counts()
print(count)

#BAR CHART
print("USING BAR CHART FOR THE FIND THE RELATIONSHIP")

count.plot(kind='bar')

plt.title("Movies vs TV Shows")
plt.ylabel("count")
plt.xlabel("type")

plt.show()





print(df.corr(numeric_only=True))
```

##Content Release Trend

- This visualization shows the distribution of Netflix content based on release year, helping identify growth patterns over time.

```python
# HISTOGRAM CHART
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("netflix_titles.csv")

count = df['type'].value_counts()
print(count)

print("How many Netflix titles were released each year.")
df['release_year'].plot(kind='hist')

plt.title("Distribution of Release Years")
plt.ylabel("count")
plt.xlabel("release year")

plt.show()
```

##Top Content Producing Countries

-This chart displays the countries contributing the highest number of titles to Netflix.

```python
#BAR CHART
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("netflix_titles.csv")

count = df['type'].value_counts()
print(count)

print("This Countries are release the most content on netflix")
df['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Count")

plt.show()
```

##Top Content Producing Countries
